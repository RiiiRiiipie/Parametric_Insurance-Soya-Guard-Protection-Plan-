```python
# Define a function to categorize rainfall
def categorize_rainfall(amount):
    if amount < 5:
        return "Water Stress"
    elif 5 <= amount <= 10:
        return "Optimal Range"
    elif 10 < amount <= 20:
        return "Excessive Rainfall"
    else:
        return "Very Heavy Rain"
df['Rainfall_Category'] = df['Daily_rainfall_mm'].apply(categorize_rainfall)
print(df)
print(df)
# Define a function to categorize temperature based on both max and min temperatures
def categorize_temperature(max_temp, min_temp):
    if max_temp < 10 or min_temp < 10:
        return "Cold Stress"
    elif 10 <= max_temp < 25 or 10 <= min_temp < 25:
        return "Cool"
    elif 25 <= max_temp <= 30 or 25 <= min_temp <= 30:
        return "Optimal Range"
    elif 30 < max_temp < 35 or 30 < min_temp < 35:
        return "Heat Stress"
    else:
        return "Extreme Heat Stress"
df['Temperature_Category'] = df.apply(lambda row: categorize_temperature(row['Daily_maximum_temperature'], row['Daily_minimum_temperature']), axis=1)
print(df)
df['Date'] = pd.to_datetime(df['DateTime'])
filtered_df = df[df['Date'].dt.month.isin(range(6, 11))]
print(filtered_df)
```

# Dividing the growth period of the soyabean in 6 different stages 
1. Germination stage which is from june to early july
2. Vegetative Stages which is from july to mid-August
3. Flowering stage which is from Mid-August to early September
4. Grain Filling stage which is from September 16 to early October
5. Harvesting Period which is from Mid-October onwards  
# Dividing the timeline in 6 different stages allow us to see in which stage the yield of soyabean is most affected by the rainfall and the temperature

```python
# Ensure the Date column is in datetime format
dfx['DateTime'] = pd.to_datetime(dfx['DateTime'])
categories = ['Germination (VE)', 'Vegetative Stages (V)', 'Flowering (R1-R2)', 
              'Pod Formation (R3-R4)', 'Grain Filling (R5-R6)', 'Harvesting Period']
dfx['Period'] = pd.Categorical([None] * len(dfx), categories=categories)
conditions = [
    (dfx['DateTime'].dt.month == 6) | ((dfx['DateTime'].dt.month == 7) & (dfx['DateTime'].dt.day <= 5)),  # Germination (VE): June to early July
    (dfx['DateTime'].dt.month == 7) | ((dfx['DateTime'].dt.month == 8) & (dfx['DateTime'].dt.day <= 15)),  # Vegetative Stages (V): July to mid-August
    ((dfx['DateTime'].dt.month == 8) & (dfx['DateTime'].dt.day > 15)) | ((dfx['DateTime'].dt.month == 9) & (dfx['DateTime'].dt.day <= 5)),  # Flowering (R1-R2): Mid-August to early September
    ((dfx['DateTime'].dt.month == 9) & (dfx['DateTime'].dt.day >= 16)) | (dfx['DateTime'].dt.month == 10),  # Pod Formation (R3-R4): September 16 to early October
    ((dfx['DateTime'].dt.month == 10) & (dfx['DateTime'].dt.day < 15)),  # Grain Filling (R5-R6): Late September to early October
    (dfx['DateTime'].dt.month >= 10) & (dfx['DateTime'].dt.day >= 15)  # Harvesting Period: Mid-October onwards
]
dfx.loc[conditions[0], 'Period'] = 'Germination (VE)'
dfx.loc[conditions[1], 'Period'] = 'Vegetative Stages (V)'
dfx.loc[conditions[2], 'Period'] = 'Flowering (R1-R2)'
dfx.loc[conditions[3], 'Period'] = 'Pod Formation (R3-R4)'
dfx.loc[conditions[4], 'Period'] = 'Grain Filling (R5-R6)'
dfx.loc[conditions[5], 'Period'] = 'Harvesting Period'
print(dfx.head(20))
```

# Obtaining the cumulative rainfall in different stages and there effects on the soaybean yield 

```python
import pandas as pd
dfx['DateTime'] = pd.to_datetime(dfx['DateTime'])
dfx['Year'] = dfx['DateTime'].dt.year
periods_of_interest = [
    'Germination Period', 
    'Vegetative Stages (V)', 
    'Flowering (R1-R2)', 
    'Pod Formation (R3-R4)', 
    'Grain Filling (R5-R6)', 
    'Harvesting Period'
]
filtered_dfx = dfx[dfx['Period'].isin(periods_of_interest)]
cumulative_rainfall = filtered_dfx.groupby(['Year', 'Period'])['Daily_rainfall_mm'].sum().reset_index()
cumulative_rainfall.rename(columns={'Daily_rainfall_mm': 'Cumulative_Rainfall'}, inplace=True)
print(cumulative_rainfall)
cumulative_rainfall.to_csv("cumulative_rainfall_by_year_and_period.csv", index=False)
```

```python
# Inputs
sum_assured = 40000
probability_of_the_events = 0.08  # Original probability (12%) # both the cases are independent there probabilities can be added 
probability_5days_cons = 32 / 3987  # 5-day consecutive rainfall > 250 mm
loading_factor = 0.2  # 20% loading # some additional charges during the insurance processs

# Calculate Expected Loss for the events (Full Sum Assured)
expected_loss_events = probability_of_the_events * sum_assured

# Calculate Expected Loss for Rainfall (50% of Sum Assured)
expected_loss_rainfall = probability_5days_cons * (sum_assured / 2)  # 50% of sum assured

# Total Expected Loss
expected_loss_total =  expected_loss_events+ expected_loss_rainfall

# Add Loading Factor
premium_with_loading = expected_loss_total * (1 + loading_factor)

# Calculate Monthly Premium
monthly_premium = premium_with_loading / 6

# Output the results
print(f"Expected Loss heavy rainfall or less rainfall: {expected_loss_events}")
print(f"Expected Loss from Rainfall: {expected_loss_rainfall}")
print(f"Total Expected Loss: {expected_loss_total}")
print(f"Premium with Loading: {premium_with_loading}")
print(f"Monthly Premium: {monthly_premium}")
```
